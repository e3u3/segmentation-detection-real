```python
import numpy as np
import pandas as pd
```

```python
original_accuacies = np.load('./Segmentation/checkpoints/unet_original_zq/accuracy_181.npy')
original_mean = np.mean(original_accuacies)
original_accuacies = np.append(original_accuacies, original_mean)

dilation_accuacies = np.load('./Segmentation/checkpoints/unet_original_zq_dilation_dilation/accuracy_180.npy')
dilation_mean = np.mean(dilation_accuacies)
dilation_accuacies = np.append(dilation_accuacies, dilation_mean)

pyramid_accuacies = np.load('./Segmentation/checkpoints/unet_original_zq_spp_spp/accuracy_180.npy')
pyramid_mean = np.mean(pyramid_accuacies)
pyramid_accuacies = np.append(pyramid_accuacies, pyramid_mean)

data = np.array([original_accuacies, dilation_accuacies, pyramid_accuacies])
columns = [f'class_{i+1}' for i in range(21)]
columns.append('mean')

df = pd.DataFrame(data, columns=columns)
df.to_csv('accuracies.csv')

df
```

```python
with open('Detection/results.txt') as f:
    results = f.readlines()
    
columns = []
aps = []
for idx, line in enumerate(results):
    elements = line.split(' ')
    if idx == len(results) - 1:
        columns.append('Mean')
    else:
        columns.append(elements[2])
    aps.append(float(elements[-1][:-2]))


df = pd.DataFrame([aps], columns=columns)
df.to_csv('precisions.csv')
pd.options.display.max_columns = 1000
   
df
```
